```python
# Import necessary libraries
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline

# we need non-linear data so we are using make_moons it will make two interleaving half circles
```

```python
# Generate a dataset of two interleaving
X,y=make_moons(n_samples=250,noise=0.05)
```

```python
# Display the shape
X
```

```python
# Plot the original dataset
plt.scatter(X[:,0],X[:,1])
```

```python
# Import StandardScaler for data normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# Initialize the StandardScaler
X_scaled=scaler.fit_transform(X)
```

```python
# Initialize DBSCAN
dbscan = DBSCAN(eps=0.5)
```

```python
# Fit DBSCAN to the scaled data
dbscan.fit(X_scaled)
```

```python
# Display the cluster labels
dbscan.labels_
```

```python
# Plot the clustered data
plt.scatter(X[:,0],X[:,1],c=dbscan.labels_)
```

```python
# Plot the original data with true labels for comparison
plt.scatter(X[:,0],X[:,1], c=y)
```
